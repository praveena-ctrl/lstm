```python
import pandas as pd
import numpy as np

# The file 'global_ai_workforce_automation_2015_2025.csv' has been uploaded.
# This code reads the CSV file into a pandas DataFrame.
df = pd.read_csv('global_ai_workforce_automation_2015_2025.csv')

# --- Initial Data Inspection ---

# 1. Display the first 5 rows to check data loading and content
print("First 5 rows of the dataset:")
print(df.head())

# 2. Display the column names, data types, and non-null counts
print("\nInformation about the dataset (column types, non-null values):")
df.info()

# 3. Display summary statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df.describe())



# Display the plot
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

# Select key numerical columns for correlation analysis
numerical_cols = [
    'AI_Investment_BillionUSD',
    'Automation_Rate_Percent',
    'Employment_Rate_Percent',
    'Average_Salary_USD',
    'Productivity_Index',
    'Reskilling_Investment_MillionUSD',
    'AI_Policy_Index',
    'Job_Displacement_Million',
    'Job_Creation_Million',
    'AI_Readiness_Score'
]
corr_df = df[numerical_cols]

# 1. Calculate the correlation matrix
correlation_matrix = corr_df.corr()

# 2. Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True, # Display the correlation values
    fmt=".2f",  # Format to two decimal places
    cmap='coolwarm', # Color map
    cbar=True,
    linewidths=.5,
    linecolor='black'
)
plt.title('Correlation Matrix of Key AI and Workforce Metrics', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()

# 3. Print the matrix for precise values
print("Correlation Matrix (Numerical Values):\n", correlation_matrix.to_string())

# 1. Calculate a new column for Net Job Change (Creation minus Displacement)
df['Net_Job_Change_Million'] = df['Job_Creation_Million'] - df['Job_Displacement_Million']

# 2. Group the data by Country and calculate the mean for three key metrics
country_summary = df.groupby('Country').agg({
    'AI_Investment_BillionUSD': 'mean',
    'Automation_Rate_Percent': 'mean',
    'Net_Job_Change_Million': 'mean'
})

# 3. Visualization: Create a grid of bar plots for country comparison

# Setup for a 3-panel plot
sns.set_style("whitegrid")
fig, axes = plt.subplots(3, 1, figsize=(14, 18))
plt.subplots_adjust(hspace=0.7)

# --- Plot 1: Average AI Investment ---
# Sort by Investment to clearly show the leaders
investment_data = country_summary.sort_values(by='AI_Investment_BillionUSD', ascending=False)
sns.barplot(x=investment_data.index, y='AI_Investment_BillionUSD', data=investment_data, ax=axes[0], palette='viridis')
axes[0].set_title('Average AI Investment (2015-2025) by Country', fontsize=16)
axes[0].set_ylabel('Average AI Investment (Billion USD)', fontsize=12)
axes[0].tick_params(axis='x', rotation=45, labelsize=10)
axes[0].set_xlabel('')

# --- Plot 2: Average Automation Rate ---
# Sort by Automation Rate
automation_data = country_summary.sort_values(by='Automation_Rate_Percent', ascending=False)
sns.barplot(x=automation_data.index, y='Automation_Rate_Percent', data=automation_data, ax=axes[1], palette='magma')
axes[1].set_title('Average Automation Rate (2015-2025) by Country', fontsize=16)
axes[1].set_ylabel('Average Automation Rate (%)', fontsize=12)
axes[1].tick_params(axis='x', rotation=45, labelsize=10)
axes[1].set_xlabel('')

# --- Plot 3: Average Net Job Change ---
# Sort by Net Job Change
net_job_data = country_summary.sort_values(by='Net_Job_Change_Million', ascending=False)
sns.barplot(x=net_job_data.index, y='Net_Job_Change_Million', data=net_job_data, ax=axes[2], palette='plasma')
axes[2].set_title('Average Net Job Change (Creation - Displacement) by Country', fontsize=16)
axes[2].set_ylabel('Average Net Job Change (Million)', fontsize=12)
axes[2].tick_params(axis='x', rotation=45, labelsize=10)
axes[2].set_xlabel('Country', fontsize=12)

plt.tight_layout()
plt.savefig('country_comparison_bar_charts.png')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

```
